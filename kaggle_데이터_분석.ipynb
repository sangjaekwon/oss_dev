#데이터 분석:
응급 상황 중 의료 콜 데이터 분석



```python
import pandas as pd

df = pd.read_csv('/content/daily_and_month_call_report.csv')
display(df.head())
```

## 데이터 시각화

### Subtask:
데이터셋 내 변수들 간의 분포와 관계를 시각화합니다.

히스토그램, 상자 그림, 산점도를 생성하여 데이터셋의 수치 변수 간의 분포와 관계를 시각화합니다.

```python
import matplotlib.pyplot as plt

numerical_columns = ['Total Number of Calls', 'Total Number of Doctors Consultancy',
                    'Number of Total Health Information', 'Number of Total Ambulance Information',
                    'Number of Total Complaints', 'Number of Calls To Know About The Service']

# 수치형 변수의 히스토그램 생성
plt.figure(figsize=(15, 10))
for i, column in enumerate(numerical_columns):
    plt.subplot(2, 3, i + 1)
    plt.hist(df[column], bins=20)
    plt.title(f"Histogram of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# 수치형 변수의 박스 플롯 생성
plt.figure(figsize=(15, 5))
plt.boxplot([df[column] for column in numerical_columns], tick_labels=numerical_columns)  # 수정된 부분
plt.title("Box Plots of Numerical Variables")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Value")
plt.show()

# 변수 간 산점도 생성
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.scatter(df['Total Number of Calls'], df['Total Number of Doctors Consultancy'])
plt.title("Total Number of Calls vs. Total Number of Doctors Consultancy")
plt.xlabel("Total Number of Calls")
plt.ylabel("Total Number of Doctors Consultancy")

plt.subplot(2, 2, 2)
plt.scatter(df['Total Number of Calls'], df['Number of Total Health Information'])
plt.title("Total Number of Calls vs. Number of Total Health Information")
plt.xlabel("Total Number of Calls")
plt.ylabel("Number of Total Health Information")

plt.subplot(2, 2, 3)
plt.scatter(df['Total Number of Calls'], df['Number of Total Ambulance Information'])
plt.title("Total Number of Calls vs. Number of Total Ambulance Information")
plt.xlabel("Total Number of Calls")
plt.ylabel("Number of Total Ambulance Information")

plt.subplot(2, 2, 4)
plt.scatter(df['Total Number of Calls'], df['Number of Total Complaints'])
plt.title("Total Number of Calls vs. Number of Total Complaints")
plt.xlabel("Total Number of Calls")
plt.ylabel("Number of Total Complaints")

plt.tight_layout()
plt.show()
```

시간(년월)에 따른 통화 카테고리의 변화를 추적하기 위해 선 그래프를 만듭니다.

```python
import matplotlib.pyplot as plt

# 시간에 따른 총 통화량 그래프
plt.figure(figsize=(12, 6))
plt.plot(df['Year'].astype(str) + '-' + df['Month'], df['Total Number of Calls'])
plt.title('Total Number of Calls Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Total Number of Calls')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 시간에 따른 다양한 통화 카테고리 그래프
plt.figure(figsize=(15, 8))

plt.plot(df['Year'].astype(str) + '-' + df['Month'], df['Total Number of Doctors Consultancy'], label='Doctors Consultancy')
plt.plot(df['Year'].astype(str) + '-' + df['Month'], df['Number of Total Health Information'], label='Health Information')
plt.plot(df['Year'].astype(str) + '-' + df['Month'], df['Number of Total Ambulance Information'], label='Ambulance Information')
plt.plot(df['Year'].astype(str) + '-' + df['Month'], df['Number of Total Complaints'], label='Complaints')
plt.plot(df['Year'].astype(str) + '-' + df['Month'], df['Number of Calls To Know About The Service'], label='Inquire About Service')

plt.title('Different Call Categories Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Number of Calls')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
